```python
import pandas as pd
import plotly
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

df_euk= pd.read_csv("../../Biostatistics/eukaryotes.csv")
df_vir= pd.read_csv("../../Biostatistics/viruses.csv")
df_pro= pd.read_csv("../../Biostatistics/prokaryotes.csv")
#df= df[df.Assembly != 'Chromosome'] #& df.Assembly == 'Scaffold']

print(df_euk.info())
print(df_euk.describe())
```

```python
df_pro.head()
```

```python
fig = px.histogram(df_euk, x="GC%")
fig.show()
```

```python
genomes = pd.DataFrame()
genomes['GC%']= df_euk['GC%']
genomes['size_mb']= df_euk['Size(Mb)']
genomes['domain']= 'eukaryote'

genomes2 = pd.DataFrame()
genomes2['GC%']= df_pro['GC%']
genomes2['size_mb']= df_pro['Size(Mb)']
genomes2['domain']= 'prokaryote'

genomes3 = pd.DataFrame()
genomes3['GC%']= df_vir['GC%']
genomes3['size_mb']= df_vir['Size(Mb)']
genomes3['domain']= 'viruses'

allgenomes= pd.concat([genomes,genomes2, genomes3], ignore_index=True)
allgenomes
```

```python
#Combined the three dataframes

fig = px.box(allgenomes, x="domain", y="GC%", color='domain')
fig.show()
```

```python
fig = px.violin(allgenomes,  x="domain", y="GC%", color='domain',
                labels={"domain": "Domain"})

fig.update_layout(showlegend=False,
    title=dict(text="%GC content of sequenced Virus, Prokaryote and Eukaryote genomes", font=dict(size=16))
)

fig.show()
fig.write_image("violin_plot.png")
```

```python
fig = px.scatter(allgenomes, y="GC%", x="size_mb", color="domain", log_x=True,
                 labels={
                     "size_mb": "Genome size (mb) \n(log scaled)",
                     "domain": "Domain"
                 })#, symbol="domain")

fig.update_layout(
    title=dict(text="%GC content of sequenced Virus, Prokaryote and Eukaryote genomes", font=dict(size=16))#, automargin=True, yref='paper')
)

fig.update_traces(marker_size=10)
fig.show()
fig.write_image("scatter_plot.png")
```
