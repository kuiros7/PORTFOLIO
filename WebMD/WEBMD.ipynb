```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
from datetime import datetime as dt
import numpy as np

df= pd.read_csv("../../webmd.csv")

#Remove empty values and na
df.dropna(subset=df.columns.difference(['Reviews']), how='all')
df= df[df.Sex != ' ']
df= df[df.Age != ' ']

#Change date to datetime format
df['Date'] = pd.to_datetime(df['Date'], format = '%m/%d/%Y')

print(df.info())
```

```python
#Choosing data, most relevant conditions
df_heat= df_new[['Condition', 'Age', 'Satisfaction']]
df_heat= df_heat.groupby(['Condition','Age']).agg(['count','mean']).reset_index()
df_heat2= df_heat.xs('Satisfaction', axis=1, drop_level=True)
df_heat= df_heat[['Condition', 'Age']]
df_heat['Size']= df_heat2['count']
df_heat['Satisfaction']= df_heat2['mean']
df_heat= df_heat[df_heat.Size > 1000 ]
df_heat= df_heat.sort_values(by='Size', ascending= False)

#Make the structure for heatmap
df_heat= df_heat[['Condition', 'Age', 'Satisfaction']]
df_heat= df_heat.groupby(['Condition', 'Age']).mean()
df_heat2= df_heat.groupby(['Condition', 'Age']).mean().reset_index()
df_heat = df_heat.unstack(level=0)

xticks_labels = df_heat2.Condition.unique()
colormap = sns.color_palette("crest", as_cmap=True)
plt.figure(figsize = (12, 5))
sns.heatmap(data =df_heat, annot = True, cbar_kws={"shrink": .5}, linewidths=.5, cmap = colormap, annot_kws={'fontsize':10})

plt.xticks(np.arange(len(xticks_labels)) + .5, labels=xticks_labels, size = 14, rotation =90)
plt.yticks(rotation=0,  size = 14)
plt.ylabel('')
plt.xlabel('')
plt.title('Score of satisfaction by condition and age group', loc='center', size= 17)

plt.savefig('heatmap_of-rate.jpg', bbox_inches='tight')
```

```python
#Grouping first info by Condition and Rate values
df_heat= df_new.groupby(['Condition']).agg(['count','mean']).reset_index()
df_heat= df_heat[['Condition', 'Satisfaction', 'Effectiveness', 'EaseofUse']]

#Obtaining count and mean values from each rate
df_heat0= df_heat.xs('Condition', axis=1, drop_level=True)
df_heat1= df_heat.xs('Satisfaction', axis=1, drop_level=True)
df_heat2= df_heat.xs('Effectiveness', axis=1, drop_level=True)
df_heat3= df_heat.xs('EaseofUse', axis=1, drop_level=True)

#Adding the info into one single dataframe that is empty
heat= pd.DataFrame(columns= ['Condition', 'Satisfaction', 'Effectiveness', 'EaseofUse'])
heat['Condition']= df_heat0
heat['Size']= df_heat1['count']
heat['Satisfaction']= df_heat1['mean']
heat['Effectiveness']= df_heat2['mean']
heat['EaseofUse']= df_heat3['mean']

#Sorting data to only get entries with the most amount of condition
heat= heat[heat.Size > 2500 ].sort_values(by='Size', ascending= False).reset_index()
heat= heat[['Condition', 'Satisfaction', 'Effectiveness', 'EaseofUse']]
heat= heat.melt(id_vars = ['Condition'], value_vars = ['Satisfaction', 'Effectiveness', 'EaseofUse'], var_name = 'Rate', value_name = 'Values')

#Make the structure for heatmap
heat= heat.groupby(['Rate', 'Condition']).mean()
heat2= heat.groupby(['Rate', 'Condition']).mean().reset_index()
heat = heat.unstack(level=0)

plt.figure(figsize = (6, 12))
colormap = sns.color_palette("crest", as_cmap=True)
sns.heatmap(data=heat, annot = True, cbar_kws={"shrink": .5}, annot_kws={'fontsize':10}, linewidths=.2, cmap = colormap, square= True)

xticks_labels = heat2.Rate.unique()
plt.xticks(np.arange(len(xticks_labels)) + .5, labels=xticks_labels, size = 14, rotation =60)
plt.yticks(rotation=0,  size = 14)
plt.ylabel('')
plt.xlabel('')
plt.title('Rate of each condition', loc='center', size= 17)

plt.savefig('heatmap_of-rate2.jpg', bbox_inches='tight')
```
