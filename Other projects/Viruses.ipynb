```python
from bs4 import BeautifulSoup
import requests

PDB_ID = "1L63"

url = requests.get("https://www.rcsb.org/fasta/entry/" + PDB_ID + "/display")
soup = BeautifulSoup(url.content, 'html.parser')
fasta_content = "".join(soup.contents)

header = fasta_content[fasta_content.index(">") + 1 : fasta_content.index("\n")]
sequence = fasta_content[fasta_content.index("\n") + 1  :]

print("Header:\n" + header)
print("Sequence:\n" + sequence)
```

```python
#https://stackoverflow.com/questions/74288454/python-webscraping-from-ncbi
#Searches by family organism
#Orthomyxoviridae 

import requests
from bs4 import BeautifulSoup
import pandas as pd

name= 'Orthoretrovirinae'
url= 'https://www.ncbi.nlm.nih.gov/Taxonomy/Browser/wwwtax.cgi?name=' + name
response = requests.get(url)
soup = BeautifulSoup(response.text)

data = []

for s in soup.select('a[title="species"]:has(+ ul a)'):
    family = s.find_previous('a',attrs={'title':'family'}).text
    genus = s.find_previous('a',attrs={'title':'genus'}).text
    
    for v in s.find_next_sibling('ul').select('a'):
        data.append({
            'virus': v.text,
            'species': s.text,
            'genus': genus,
            'family': family
        })

pd.DataFrame(data)
```
